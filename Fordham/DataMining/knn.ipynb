# KNN

```python
# Import required libraries
import numpy as np
import pandas as pd
```

```python
# Read csv files
train_set = pd.read_csv('train.csv', index_col="id")
test_set = pd.read_csv('test.csv', index_col="id")
# Create X_train, y_train, X_test variables
X_train = train_set.drop("CLASS_LABEL", axis=1)
y_train = train_set["CLASS_LABEL"]
X_test = test_set.copy()
```

```python
# Defining KNN function
def KNN(X_train, y_train, X_test,k=3):
    # List to store results
    predictions = []
    for row in range(0, len(X_test)):
        # Euclidean function
        euclidean_distance = np.sum((X_test.iloc[row] - X_train)**2,axis=1)**(1/2)
        euclidean_distance = euclidean_distance.rename("Euclidean_distance")
        # Determine highest voted class
        df = pd.merge(y_train,euclidean_distance, on= "id").sort_values("Euclidean_distance").iloc[:k]
        df["Count"] = 1
        df = df.groupby("CLASS_LABEL").sum().sort_values("Count", ascending=False)
        max_count = max(df.Count)
        # If two or more winning classes receive the same number of votes, we break the tie using the lowest total distance
        if len(df[df["Count"] == max_count]) > 1:
            df = df[df["Count"] == max_count].sort_values("Euclidean_distance")
            prediction = df.sort_values("Euclidean_distance").index[0]
        else:
            prediction = df.index[0]

        predictions.append(prediction)

    return np.array(predictions)
```

```python
# Inserting predictions into new columns
X_test["Preds_k1"] = KNN(train_set.drop("ExampleID",axis=1), y_train, test_set.drop("ExampleID",axis=1),k=1)
X_test["Preds_k3"] = KNN(train_set.drop("ExampleID",axis=1), y_train, test_set.drop("ExampleID",axis=1),k=3)
X_test["Preds_k5"] = KNN(train_set.drop("ExampleID",axis=1), y_train, test_set.drop("ExampleID",axis=1),k=5)
X_test["Preds_k7"] = KNN(train_set.drop("ExampleID",axis=1), y_train, test_set.drop("ExampleID",axis=1),k=7)
X_test["Preds_k9"] = KNN(train_set.drop("ExampleID",axis=1), y_train, test_set.drop("ExampleID",axis=1),k=9)
```
